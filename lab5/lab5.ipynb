```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

plt.plot(x, y, label = 'sin(x)')
plt.plot(x, z, label = 'cos(x)')
plt.title(' 1   ')
plt.xlabel(' x')
plt.ylabel(' y')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


mas = np.random.normal(size=1000)
plt.hist(mas, bins=20)
plt.title(' 2   ')
plt.xlabel(' x')
plt.ylabel(' y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


matrix = np.random.randint(1, 11, (5, 5))
print(matrix)
print(' :')
print(np.mean(matrix))
print(' :')
print(np.max(matrix))
print(' :')
print(np.min(matrix))
print('  :')
print(np.sum(matrix, axis=0))
print()
```

```python
import matplotlib.pyplot as plt
import numpy as np


matrix = np.random.randint(1, 11, (5, 5))
plt.imshow(matrix)
for y in range(0, 5):
  for x in range(0, 5):
    plt.text(x, y, matrix[x, y])
plt.colorbar()
plt.title(' 4   ')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 10, 100)
plt.subplot(5, 1, 1)
plt.title(' ')
plt.plot(x, x**2)

x = np.random.randint(1, 100, 25)
y = np.random.randint(1, 100, 25)
plt.subplot(5, 1, 3)
plt.title('   ')
plt.scatter(x, y)

titles = ['A', 'B', 'C']
mas = [3, 7, 2]
plt.subplot(5, 1, 5)
plt.title('    ')
plt.bar(titles, mas)

plt.show()
```
