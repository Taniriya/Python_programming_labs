```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

plt.plot(x, y, x, z)
plt.title('Задание 1 Посторение простых графиков')
plt.legend()
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


mas = np.random.normal(size=1000)
plt.hist(mas, bins=20)
plt.title('Задание 2 Гистограммы случайных данных')
plt.legend()
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


matrix = np.random.randint(1, 11, (5, 5))
print(matrix)
print('Среднее значение:')
print(np.mean(matrix))
print('Максимальный элемент:')
print(np.max(matrix))
print('Минимальный элемент:')
print(np.min(matrix))
print('Сумма по столбцам:')
print(np.sum(matrix, axis=0))
print()
```

```python
import matplotlib.pyplot as plt
import numpy as np


matrix = np.random.randint(1, 11, (5, 5))
plt.imshow(matrix)
for y in range(0, 5):
  for x in range(0, 5):
    plt.text(x, y, matrix[x, y])
plt.colorbar()
plt.title('Задание 4 Визуализация тепловой карты')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 10, 100)
plt.subplot(5, 1, 1)
plt.title('Линейный график')
plt.plot(x, x**2)

x = np.random.randint(1, 100, 25)
y = np.random.randint(1, 100, 25)
plt.subplot(5, 1, 3)
plt.title('Точечный график случайных точек')
plt.scatter(x, y)

titles = ['A', 'B', 'C']
mas = [3, 7, 2]
plt.subplot(5, 1, 5)
plt.title('Столбчатая диаграмма для категориальных данных')
plt.bar(titles, mas)

plt.show()
```
